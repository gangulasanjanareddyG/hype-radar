```python
# ============================================================
# CELL PURPOSE: Load cleaned dataset
# ============================================================

import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np

DATA_PATH = "/Users/sanjana/Desktop/Hype-Predictor/Data"

df = pd.read_csv(f"{DATA_PATH}/clean_data.csv")

print(df)
```

```python
# ============================================================
# CELL PURPOSE: Check scale before normalization
# ============================================================

print(df.describe())
```

```python
# ============================================================
# CELL PURPOSE: Normalize signals using Z-score
# ============================================================

cols = ["trends_score", "reddit_score", "youtube_score", "news_score"]

df_z = df.copy()

for col in cols:
    mean = df[col].mean()
    std = df[col].std()
    df_z[col] = (df[col] - mean) / std

print(df_z)
```

```python
# ============================================================
# CELL PURPOSE: Convert Z-scores to 0–100 scale
# ============================================================

df_scaled = df_z.copy()

for col in cols:
    min_val = df_z[col].min()
    max_val = df_z[col].max()
    df_scaled[col] = 100 * (df_z[col] - min_val) / (max_val - min_val)

print(df_scaled)
```
